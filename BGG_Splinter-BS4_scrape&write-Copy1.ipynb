```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
 
    return None
gamePrices = {}


bggSQLdb = create_connection("database.sqlite/database.sqlite")
def process_all_games(conn):
    global gamePrices
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    cur.execute("SELECT \"game.id\", \"details.name\" FROM BoardGames WHERE \"price.done\" is null") #  IS NULL 
    
    rows = cur.fetchall()
    conn.close()
    return rows


def process_games(rows, conn):
    bggSQLdb = create_connection("database.sqlite/database.sqlite")
    cur2 = conn.cursor()
    gameNumber = 0
    
    for row in rows:
        print(row[1] + " :: " + row[0])
        url = f'https://boardgamegeek.com/boardgame/{row[0]}/'
        browser.visit(url)
        print(url)
        html = browser.html
        soup = BeautifulSoup(html, 'html.parser')
        sales = soup.find_all("div", {"class":"summary-sale-item-price"}) #summary-sale-item-price-retail (for suggested retail)
        thisPrices = []
        #gamePrices[row[0]] ={}
        gamePrices[row[0]] = {'Name':row[1]}
        for price in sales:
            #print(price.text.find('$'))
            if price.text.find('$') == 1:
                #print('found!')
                usPrice = price.text[2:-1]
                print(usPrice)
                thisPrices.append(usPrice)
        gamePrices[row[0]]['Prices'] = thisPrices
        print(gamePrices[row[0]])
        
        gamePrice = ', '.join(thisPrices)
        gameId = row[0]
        cur2 = bggSQLdb.cursor()
        print(f"update BoardGames set \"price.done\" = '{gameId}',  \"price.prices\" = '{gamePrice}'  where \"game.id\" = '{gameId}'")
        cur2.execute(f"update BoardGames set \"price.done\" = '{gameId}',  \"price.prices\" = '{gamePrice}'  where \"game.id\" = '{gameId}'")
        rows2 = cur2.fetchall()
        print(rows2)
        print(bggSQLdb.commit())
        #time.sleep(.11)
        gameNumber = gameNumber +1
        
        
bggSQLdb = create_connection("database.sqlite/database.sqlite")
bigG = process_all_games(bggSQLdb)
bggSQLdb.close()
bggSQLdb = create_connection("database.sqlite/database.sqlite")
process_games(bigG, bggSQLdb)
bggSQLdb.close()
```
