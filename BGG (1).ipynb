```python
import pandas as pd
import pymysql
pymysql.install_as_MySQLdb()

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
Base = automap_base()
engine = create_engine("sqlite:///Resources/database.sqlite", execution_options={'sqlite_raw_colnames':True})
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
conn = engine.connect()
inspector = inspect(engine)
inspector.get_table_names()
```

```python
columns = inspector.get_columns('BoardGames')
for c in columns:
    print(c['name'], c['type'])
```

```python
game_details=pd.read_sql('SELECT [game.id] , [details.name] , [details.yearpublished], [attributes.boardgamepublisher], [game.type], [details.description], [details.minage], [details.minplayers], [details.maxplayers], [details.playingtime],[details.minplaytime], [details.maxplaytime]FROM BoardGames', conn)
game_details.head()
```

```python
#pd.read_sql('SELECT [game.id], [details.name] FROM BoardGames WHERE [details.name] LIKE '%expansion%'', conn)
game_details.columns = game_details.columns.str.strip().str.lower().str.replace('.','_')
game_details
```

```python
game_details.game_type.unique()
```

```python
game_details.game_type.count()
```

```python
clean_game_details = game_details[game_details.game_type != 'boardgameexpansion']
clean_game_details.game_type.unique()
```

```python
clean_game_details.game_type.count()
```

```python
clean_game_details=clean_game_details.rename(columns={'game_id':'id',
                                                      'details_name':'name',
                                                      'details_yearpublished':'yearpublished',
                                                      'attributes_boardgamepublisher':'boardgamepublisher',
                                                      'game_type':'gametype',
                                                      'details_description':'descriptions',
                                                      'details_minage':'minage',
                                                      'details_minplayers':'minplayers',
                                                      'details_maxplayers':'maxplayers',
                                                      'details_playingtime':'playtime',
                                                      'details_minplaytime':'minplaytime',
                                                      'details_maxplaytime':'maxplaytime'})
```

```python
clean_game_details.head()
```
